```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("housing.csv")
data.head()
```

```python
data.head(10)
```

```python
data.describe(include = "all")
```

```python
data["ocean_proximity"]
```

```python

data_pie = data.groupby("ocean_proximity").size()
data_pie.plot(kind='pie', subplots=True, figsize=(8, 8))
plt.title("Pie Chart of Ocean Proximity")
plt.ylabel("")
plt.show()
```

```python
import random


c = ["green", "red", "yellow", "blue", "cyan"]

types = data["ocean_proximity"].unique()
data_group = data.groupby("ocean_proximity")
values = []
for type in types:
    values.append(data_group.mean()["median_house_value"][type])
plt.bar(types, values, color = c)
plt.show()
```

```python
import random


c = ["green", "red", "yellow", "blue", "cyan"]

types = data["housing_median_age"].unique()
types.sort()
data_group = data.groupby("housing_median_age")
values = []
for type in types:
    values.append(data_group.mean()["median_income"][type])
types = data["housing_median_age"]
values = data["median_income"]
plt.scatter(types, values)
plt.show()
```
